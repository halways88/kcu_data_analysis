```python
import pandas as pd
```

```python
listings = pd.read_csv("listings.csv")

```

```python
listings = listings[['name', 'id', 'latitude', 'longitude', 'property_type', 'room_type', 'bathrooms', 'bedrooms', 'amenities', 'price', 'cleaning_fee', 'cancellation_policy', 'reviews_per_month', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']]
```

```python
modified_amenities = []
for item in list(listings["amenities"]):
    modified_item = item[1:-1].replace("\"","").split(",")
    modified_amenities.append(modified_item)

listings["amenities"] = modified_amenities
listings
```

```python
amenities_list = []
for items in listings["amenities"]:
    for item in items:
        if item in amenities_list:
            pass
        elif item not in amenities_list:
            amenities_list.append(item)
amenities_list
```
